## 📖 Concept Explanation

### What is Feature Engineering?

**Feature Engineering** is the process of using your existing data to create **new columns (features)** that make your dataset more informative and useful for analysis or machine learning.

Think of it like this:

| Real World Analogy | Feature Engineering Equivalent |
|---|---|
| You know someone's birthday → You calculate their **age** | You have a Date column → You extract the **Month** |
| You know price and quantity → You calculate **total cost** | You have Price and Quantity → You create **Revenue per Unit** |
| You see a bill amount → You say it's "**expensive**" or "**cheap**" | You have Revenue → You create a **Revenue Category** (Low/Medium/High) |

### Why is Feature Engineering Important?

Your raw data often doesn't tell the full story. Feature Engineering helps you:

- 📊 **Discover hidden patterns** — Which day of the week gets most sales?
- 🎯 **Improve analysis** — Are high-quantity orders more profitable per unit?
- 🤖 **Prepare data for ML** — Machine learning models need numerical, well-structured features
- 💡 **Answer business questions** — What percentage of orders are "high value"?

### Types of Feature Engineering

| Type | What it does | Example |
|---|---|---|
| **Date Features** | Extract parts of a date | Date → Year, Month, Day, Day Name |
| **Mathematical Features** | Combine numbers with math | Revenue / Quantity = Revenue per Unit |
| **Categorical Features** | Group numbers into categories | Revenue → Low, Medium, High |
| **Boolean/Flag Features** | Create True/False columns | Is the order above ₹1000? Yes/No |
| **Encoding Features** | Convert text to numbers | Electronics → 1, Accessories → 0 |

### Step 2: Date Feature Engineering 📅

The **Date** column is currently stored as a **string** (text). We need to:
1. **Convert** it to datetime format
2. **Extract** useful parts like Year, Month, Day, and Day of Week

#### Why convert to datetime?
- Strings like `"2026-01-07"` are just text — Python doesn't know it's a date
- Datetime format lets us **extract parts**, **sort by date**, and **calculate differences**

```python
# ----- Convert 'Date' column from string to datetime -----

# Before conversion — check the data type
print("BEFORE conversion:")
print(f"  Date column type: {df['Date'].dtype}")
print(f"  Sample value: {df['Date'].iloc[0]}")
print(f"  Type of value: {type(df['Date'].iloc[0])}")
print()

# Convert the 'Date' column to datetime format
# pd.to_datetime() is a powerful function that can parse many date formats
df['Date'] = pd.to_datetime(df['Date'])

# After conversion — verify the change
print("AFTER conversion:")
print(f"  Date column type: {df['Date'].dtype}")
print(f"  Sample value: {df['Date'].iloc[0]}")
print(f"  Type of value: {type(df['Date'].iloc[0])}")
print()
print("✅ Date column converted to datetime successfully!")
```

```python
# ----- Extract Year from Date -----

# .dt is a special accessor for datetime columns
# .dt.year gives us just the year part of the date

df['Order_Year'] = df['Date'].dt.year

# Let's see the result
print("✅ 'Order_Year' column created!")

print(df[['Date', 'Order_Year']].head())
```

```python
# ----- Extract Month from Date -----

# .dt.month gives the month number (1 = January, 2 = February, etc.)
df['Order_Month'] = df['Date'].dt.month

# .dt.month_name() gives the full month name (January, February, etc.)
df['Order_Month_Name'] = df['Date'].dt.month_name()

# Let's see both
print("✅ 'Order_Month' and 'Order_Month_Name' columns created!")
print(df[['Date', 'Order_Month', 'Order_Month_Name']].head())
```

```python
# ----- Extract Day and Day of Week from Date -----

# .dt.day gives the day of the month (1-31)
df['Order_Day'] = df['Date'].dt.day

# .dt.day_name() gives the name of the day (Monday, Tuesday, etc.)
df['Day_Name'] = df['Date'].dt.day_name()

# .dt.dayofweek gives a number (0 = Monday, 6 = Sunday)
df['Day_of_Week'] = df['Date'].dt.dayofweek

# Let's see all the date features together
print("✅ Day features created!")
print()
print(df[['Date', 'Order_Day', 'Day_Name', 'Day_of_Week']].head(10))
```

#### 📋 Summary of `.dt` accessor methods:

| Code | What it extracts | Example Output |
|---|---|---|
| `df['Date'].dt.year` | Year | 2026 |
| `df['Date'].dt.month` | Month number | 1 (January) |
| `df['Date'].dt.month_name()` | Month name | January |
| `df['Date'].dt.day` | Day of month | 7 |
| `df['Date'].dt.day_name()` | Day name | Wednesday |
| `df['Date'].dt.dayofweek` | Day number (0=Mon) | 2 (Wednesday) |
| `df['Date'].dt.quarter` | Quarter | 1 (Q1) |
| `df['Date'].dt.week` | Week number | 2 |

> 💡 **Remember**: The `.dt` accessor ONLY works on datetime columns. If your column is a string, convert it first with `pd.to_datetime()`!

### Step 3: Mathematical Feature Engineering 🔢

Now let's create new features by doing **math** on existing columns.

This is like a shopkeeper calculating:
- "How much did I earn **per item** sold?"
- "What **percentage** of revenue is my profit?"

```python
# ----- Feature: Revenue per Unit -----

# Revenue per Unit = Total Revenue / Quantity
# This tells us how much revenue each single unit generated

df['Revenue_Per_Unit'] = df['Revenue'] / df['Quantity']

# Let's see the result
print("✅ 'Revenue_Per_Unit' column created!")
print()
print(df[['Product', 'Price', 'Quantity', 'Revenue', 'Revenue_Per_Unit']].sample(10))
```

```python
# ----- Feature: Total Cost (simulated) -----

# In real business, Cost = Price × some cost factor
# Let's assume the cost is 60% of the selling price (40% margin)
# This is a common practice when actual cost data isn't available

cost_percentage = 0.60 # 60% of price is the cost

df['Cost'] = df['Price'] * cost_percentage * df['Quantity']

print("✅ 'Cost' column created! (Assuming cost = 60% of selling price)")
print()
print(df[['Product', 'Price', 'Quantity', 'Revenue', 'Cost']].head())
```

```python
# ----- Feature: Profit -----

# Profit = Revenue - Cost
# This tells us how much money we actually made on each transaction

df['Profit'] = df['Revenue'] - df['Cost']

print("✅ 'Profit' column created!")
print()
print(df[['Product', 'Revenue', 'Cost', 'Profit']].sample(10))
```

```python
# ----- Feature: Profit Margin (%) -----

# Profit Margin = (Profit / Revenue) × 100
# This tells us what percentage of revenue is actual profit

df['Profit_Margin'] = (df['Profit'] / df['Revenue']) * 100

# Round to 2 decimal places for clean output
df['Profit_Margin'] = df['Profit_Margin'].round(2)

print("✅ 'Profit_Margin' column created!")
print()
print(df[['Product', 'Revenue', 'Profit', 'Profit_Margin']].head(10))
print()
print(f"📊 Average Profit Margin: {df['Profit_Margin'].mean():.2f}%")
```

### Step 4: Categorical Feature Engineering 🏷️

Sometimes numbers are more useful when we put them into **groups** or **categories**.

For example:
- Revenue of ₹300 → "**Low Revenue**"
- Revenue of ₹1500 → "**Medium Revenue**"
- Revenue of ₹4000 → "**High Revenue**"

We'll learn two methods:
1. **`pd.cut()`** — for cutting data into fixed bins
2. **`np.where()`** — for simple if/else conditions

```python
# ----- Feature: Revenue Category using pd.cut() -----

# pd.cut() divides a numerical column into bins (groups)
# We define the boundaries: 0-500 = Low, 500-2000 = Medium, 2000-5000 = High

# Define the bin edges
bins = [0, 500, 2000, 5000]

# Define the labels for each bin
labels = ['Low', 'Medium', 'High']

# Create the new column
df['Revenue_Category'] = pd.cut(
    df['Revenue'],       # Column to categorize
    bins=bins,           # The boundary values
    labels=labels        # Labels for each group
)

print("✅ 'Revenue_Category' column created!")
print()
print(df[['Product', 'Revenue', 'Revenue_Category']].head(10))
print()
print("📊 Distribution of Revenue Categories:")
print(df['Revenue_Category'].value_counts())
```

```python
# ----- Feature: Quantity Category using np.where() -----

# np.where() works like an IF-ELSE statement
# Syntax: np.where(condition, value_if_true, value_if_false)

# Let's classify orders:
# If Quantity >= 3 → "Bulk Order"
# Otherwise → "Regular Order"

df['Order_Type'] = np.where(
    df['Quantity'] >= 3,      # Condition
    'Bulk Order',             # Value if True
    'Regular Order'           # Value if False
)

print("✅ 'Order_Type' column created!")
print()
print(df[['Product', 'Quantity', 'Order_Type']].head(10))
print()
print("📊 Distribution:")
print(df['Order_Type'].value_counts())
```

```python
# ----- Feature: Price Tier using multiple conditions with np.select() -----

# What if we have MORE than 2 categories?
# np.where() only handles 2 (if/else)
# np.select() can handle multiple conditions!

# Define the conditions
conditions = [
    df['Price'] <= 200,                          # Condition 1: Budget
    (df['Price'] > 200) & (df['Price'] <= 600),  # Condition 2: Mid-Range
    df['Price'] > 600                             # Condition 3: Premium
]

# Define the values for each condition
choices = ['Budget', 'Mid-Range', 'Premium']

# Create the column
df['Price_Tier'] = np.select(conditions, choices, default='Unknown')

print("✅ 'Price_Tier' column created!")
print()
print(df[['Product', 'Price', 'Price_Tier']].drop_duplicates().sort_values('Price'))
print()
print("📊 Distribution:")
print(df['Price_Tier'].value_counts())
```

#### 📋 When to use which method?

| Method | Best For | Example |
|---|---|---|
| `np.where()` | **2 categories** (if/else) | High Value: Yes / No |
| `np.select()` | **3+ categories** (if/elif/else) | Budget / Mid-Range / Premium |
| `pd.cut()` | **Equal-width bins** on numbers | Low / Medium / High |
| `pd.qcut()` | **Equal-frequency bins** | Bottom 25% / Middle 50% / Top 25% |
| `.map()` | **Mapping exact values** | Product → Category |
| `.apply()` | **Complex custom logic** | Custom function on each row |

### Step 5: Boolean (Flag) Feature Engineering 🚩

Boolean features are **True/False** columns that act as flags or indicators.

They are useful for:
- Quickly filtering data
- Counting how many rows meet a condition
- Using as features in machine learning

```python
# ----- Feature: Is Weekend? -----

# Day_of_Week: 0=Monday, 1=Tuesday, ..., 5=Saturday, 6=Sunday
# Weekend means Day_of_Week is 5 (Saturday) or 6 (Sunday)

df['Is_Weekend'] = df['Day_of_Week'] >= 5

print("✅ 'Is_Weekend' column created!")
print()
print(df[['Date', 'Day_Name', 'Day_of_Week', 'Is_Weekend']].head(10))
print()
print(f"📊 Weekend orders: {df['Is_Weekend'].sum()} out of {len(df)}")
print(f"   Weekday orders: {(~df['Is_Weekend']).sum()} out of {len(df)}")
```

```python
# ----- Feature: Is High Value Order? -----

# Let's flag orders where Revenue is above ₹1000

df['Is_High_Value'] = df['Revenue'] > 1000

print("✅ 'Is_High_Value' column created! (Revenue > ₹1000)")
print()
print(df[['Product', 'Revenue', 'Is_High_Value']].head(10))
print()
print(f"📊 High value orders: {df['Is_High_Value'].sum()} ({df['Is_High_Value'].mean()*100:.1f}%)")
print(f"   Regular orders:    {(~df['Is_High_Value']).sum()} ({(~df['Is_High_Value']).mean()*100:.1f}%)")
```

```python
# ----- Feature: Is Multi-Unit Order? -----

# Flag orders where more than 1 unit was purchased

df['Is_Multi_Unit'] = df['Quantity'] > 1

print("✅ 'Is_Multi_Unit' column created! (Quantity > 1)")
print()
print(df[['Product', 'Quantity', 'Is_Multi_Unit']].head(10))
print()
print(f"📊 Multi-unit orders: {df['Is_Multi_Unit'].sum()} ({df['Is_Multi_Unit'].mean()*100:.1f}%)")
```

> 💡 **Pro Tip**: Boolean columns are secretly 0s and 1s!
> - `True` = 1, `False` = 0
> - So `df['Is_Weekend'].sum()` counts how many `True` values there are
> - And `df['Is_Weekend'].mean()` gives the proportion (percentage) of `True` values

### Step 6: Encoding Categorical Variables 🔄

Machine Learning models and many analysis techniques need **numbers**, not text. So we need to convert categorical (text) columns into numbers.

Two common methods:
1. **Label Encoding** — Convert each category to a unique number
2. **One-Hot Encoding** — Create a separate column for each category with 0/1 values

```python
# ----- Method 1: Label Encoding using .map() -----

# We manually map each category to a number
# This works well when there are only a few categories

# Our dataset has 2 categories: Electronics and Accessories
category_mapping = {
    'Electronics': 1,
    'Accessories': 0
}

# Apply the mapping
df['Category_Encoded'] = df['Category'].map(category_mapping)

print("✅ 'Category_Encoded' column created using Label Encoding!")
print()
print(df[['Category', 'Category_Encoded']].drop_duplicates())
print()
print("Mapping used:")
for key, value in category_mapping.items():
    print(f"  {key} → {value}")
```

```python
# ----- Method 2: One-Hot Encoding using pd.get_dummies() -----

# One-Hot Encoding creates a NEW column for EACH unique category
# Each column has 1 if that row belongs to that category, 0 otherwise

# Let's one-hot encode the 'Product' column
product_dummies = pd.get_dummies(
    df['Product'],          # Column to encode
    prefix='Product'        # Prefix for new column names
)

print("✅ One-Hot Encoding of 'Product' column:")
print()
print(product_dummies.head(10))
print()
print(f"📊 Original 'Product' column had {df['Product'].nunique()} unique values")
print(f"   One-Hot Encoding created {product_dummies.shape[1]} new columns")
```

```python
# ----- Add the one-hot encoded columns to our DataFrame -----

# Use pd.concat() to join the dummy columns with our main DataFrame
df = pd.concat([df, product_dummies], axis=1)

print("✅ One-hot encoded columns added to DataFrame!")
print()
print(f"DataFrame now has {df.shape[1]} columns (up from 7 original columns!)")
print()
print("New columns added:")
for col in product_dummies.columns:
    print(f"  • {col}")
```

#### 📋 Label Encoding vs One-Hot Encoding

| Aspect | Label Encoding | One-Hot Encoding |
|---|---|---|
| **How it works** | Assigns a number to each category | Creates a column per category |
| **Example** | Electronics=1, Accessories=0 | Product_Laptop=1, Product_Tablet=0 |
| **Best for** | 2 categories or ordered categories | 3+ unordered categories |
| **Risk** | Model may think 1 > 0 (order matters) | Creates many columns if many categories |
| **Method** | `.map()` or `LabelEncoder` | `pd.get_dummies()` |

### Step 7: Using `.apply()` for Custom Features 🛠️

Sometimes you need **custom logic** that's too complex for `np.where()`. In that case, use `.apply()` with a custom function.

```python
# ----- Feature: Revenue Label using a custom function with .apply() -----

# Define a function that takes a revenue value and returns a label
def classify_revenue(revenue):
    """Classify revenue into descriptive labels."""
    if revenue >= 3000:
        return '🟢 Excellent'
    elif revenue >= 1500:
        return '🟡 Good'
    elif revenue >= 500:
        return '🟠 Average'
    else:
        return '🔴 Low'

# Apply the function to each value in the 'Revenue' column
df['Revenue_Label'] = df['Revenue'].apply(classify_revenue)

print("✅ 'Revenue_Label' column created using .apply()!")
print()
print(df[['Product', 'Revenue', 'Revenue_Label']].head(10))
print()
print("📊 Distribution:")
print(df['Revenue_Label'].value_counts())
```

```python
# ----- Feature: Using .apply() with lambda (short functions) -----

# Lambda functions are one-line shortcut functions
# Great for simple transformations

# Example: Create a 'Transaction_Number' by extracting the number from Transaction_ID
# 'TXN-1001' → 1001
import pandas as pd

df=pd.read_excel("DataSet_used/sales_data_messy.xlsx")
df=df.dropna(subset="Transaction_ID")
print(df["Transaction_ID"].isnull().sum().sum())
print(df["Transaction_ID"].dtype)
df['Transaction_Number'] =(df['Transaction_ID'].apply(lambda x:(x.split("-"))))

print("✅ 'Transaction_Number' extracted using lambda + .apply()!")
print()
print(df[['Transaction_ID', 'Transaction_Number']].head())
print()
print("How it works:")
print("  'TXN-1001'.split('-') → ['TXN', '1001']")
print("  ['TXN', '1001'][1]   → '1001'")
print("  int('1001')          → 1001")
```
