```python
# Task 1: Import libraries and load the dataset
import pandas as pd
import numpy as np
df = pd.read_csv( "test_dataset.csv")

print("Dataset loaded successfully")
```

```python
# Task 2: Basic data inspection
# print(df.head())
# print(df.tail())
# print("Rows and columns:", df.shape)
# print("Column names:", df.columns.tolist())
print(df.dtypes)
# df.info()
# print(df.describe())
```

```python
# Task 3: Check missing values
missing_count = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

missing_summary = pd.DataFrame({
    "Missing_Count": missing_count,
    "Missing_Percentage": missing_percentage
})

print(missing_summary)
print("Columns with missing values:")
print(missing_summary[missing_summary["Missing_Count"] > 0])
```

```python
# Convert numeric columns to proper numeric data type
numeric_columns = ["Discount","Sales","Delivery_Days","Rating"]

for column in numeric_columns:
    if column in df.columns:
        df[column] = pd.to_numeric(df[column], errors="coerce")

print(df.dtypes)
```

```python
# Task 4: Handle missing values
if "Transaction_ID" in df.columns:
    df["Transaction_ID"] = df["Transaction_ID"].fillna("Unknown")

text_columns = df.select_dtypes(include="object").columns

for column in text_columns:
    if df[column].isnull().any():
        df[column] = df[column].fillna(df[column].mode()[0])

numeric_columns = df.select_dtypes(include="number").columns

for column in numeric_columns:
    if df[column].isnull().any():
        df[column] = df[column].fillna(df[column].median())

print(df.isnull().sum())
```

```python
# Task 5 and 6: Check and remove duplicate rows
print("Duplicate rows:", df.duplicated().sum())

df = df.drop_duplicates().reset_index(drop=True)

print("Duplicate rows after removal:", df.duplicated().sum())
```

```python
# Task 8: Clean text columns
text_columns = df.select_dtypes(include="object").columns

for column in text_columns:
    df[column] = df[column].astype("string").str.strip().str.title()


print(df.head())
```

```python
# Task 9: Check unique values
columns_to_check = [
"Category",
"Returned",
"Payment_Mode",
"Customer_Name"
]

for column in columns_to_check:
        print("\n", column)
        print(df[column].unique())
```

```python
# Task 10: Convert the date column
df["Order_Date"] = pd.to_datetime(df["Order_Date"], errors="coerce",format="mixed")

print("Invalid dates:", df["Order_Date"].isnull().sum())

print(df["Order_Date"].head())
```

```python
# Task 14: Handle invalid values
df.loc[df["Quantity"] < 0, "Quantity"] = np.nan

if "Discount" in df.columns:
    df.loc[(df["Discount"] < 0) | (df["Discount"] > 1), "Discount"] = np.nan

if "Rating" in df.columns:
    df.loc[(df["Rating"] < 0) | (df["Rating"] > 5), "Rating"] = np.nan

columns_to_fill = ["Price", "Quantity", "Discount", "Rating"]

for column in columns_to_fill:
    if column in df.columns:
        df[column] = df[column].fillna(df[column].median())

print("Invalid values handled")
```

```python
# Task 15: Detect outliers using the IQR method
outlier_columns = ["Quantity", "Discount", "Rating","Sales"]
# Select the column

for column in outlier_columns:
    # Step 1: Calculate Q1 and Q3
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    # Step 2: Calculate IQR
    IQR = Q3 - Q1

    # Step 3: Calculate lower and upper limits
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    print(f"Lower Limit of {column}", lower_limit)
    print(f"Upper Limit of {column}", upper_limit)

    #Detection
    outliers = df[
    (df[column] < lower_limit) |
    (df[column] > upper_limit)]
    print("Number of outliers:", len(outliers))
    print()
   
```

```python
#Handling Outliers
outlier_columns = ["Quantity", "Discount", "Rating","Sales"]

for column in outlier_columns:
   

    df[column] = df[column].clip(
    lower=lower_limit,
    upper=upper_limit)

    #checking after updation
    outliers_after_capping = df[
    (df[column] < lower_limit) |
    (df[column] > upper_limit)]

    print("Outliers after capping:", len(outliers_after_capping))
```

```python
# Task 18: Create new columns

df["Total_Amount"] = df["Sales"] * df["Quantity"]
df["Discount_Amount"] = df["Total_Amount"] * df["Discount"]
df["Final_Amount"] = df["Total_Amount"] - df["Discount_Amount"]

df["Order_Year"] = df["Order_Date"].dt.year
df["Order_Month"] = df["Order_Date"].dt.month_name()
df["Order_Day"] = df["Order_Date"].dt.day

print(df.head())
```

```python
# Part I: Analysis questions

# 1. Total sales amount
total_sales = df["Sales"].sum()
print("Total sales amount:", total_sales)

# 2. Category with the highest total sales
category_sales = df.groupby("Category")["Sales"].sum()
print("\nSales by category:")
print(category_sales)
print("Highest sales category:", category_sales.idxmax())

# 4. Product with the highest quantity sold
product_quantity = df.groupby("Product_Name")["Quantity"].sum()
print("\nQuantity by product:")
print(product_quantity)
print("Product with highest quantity:", product_quantity.idxmax())

# 5. Average rating
print("\nAverage rating:", df["Rating"].mean())
```

```python
# More analysis questions

# 6. Most-used payment method

print("Most-used payment method:", df["Payment_Mode"].mode()[0])

# 7. Month with the highest sales
month_sales = df.groupby("Order_Month")["Sales"].sum()
print("\nSales by month:")
print(month_sales)
print("Month with highest sales:", month_sales.idxmax())

# 8. Top 5 highest sales records
print("\nTop 5 highest sales records:")
df=df.sort_values("Sales",ascending=False)
print("top 5 sales values:\n",df["Sales"].head())
# 9. Bottom 5 sales records
print("\nBottom 5 sales records:")
print("bottom 5 sales values:\n",df["Sales"].tail())

# 10. Number of orders in each category
print("\nOrders in each category:")
print(df["Category"].value_counts())
```

```python
# Part J: GroupBy practice

# 1. Total sales by category
print("Total sales by category:")
print(df.groupby("Category")["Sales"].sum())

# 3. Average rating by category

print("\nAverage rating by category:")
print(df.groupby("Category")["Rating"].mean())

# 4. Total quantity sold by product
print("\nTotal quantity sold by product:")
print(df.groupby("Product_Name")["Quantity"].sum())

# 5. Average final amount by payment method

print("\nAverage final amount by payment method:")
print(df.groupby("Payment_Mode")["Final_Amount"].mean())
```

```python
# Task 20: Save the cleaned dataset
output_file = "sales_data_cleaned.xlsx"
df.to_excel(output_file, index=False)

print("Cleaned dataset saved as:", output_file)
```
