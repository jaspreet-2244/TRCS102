```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
sns.set_palette('Set2')
print(sns.__version__)
```

```python
df=pd.read_excel("DataSet_used/sales_data.xlsx")
print("shape of dataset:",df.shape)
df.head(5)
```

```python
# countplot use
plt.figure(figsize=(10,5))
sns.countplot(x='Product',data=df,order=df["Product"].value_counts().index,edgecolor='black')
plt.xlabel("Product",fontsize=10)
plt.ylabel("Number of Transections ", fontsize=10)
plt.title("Transaction count by Product",color='Pink',fontsize=20)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df , x="Category",order=df["Category"].value_counts().index,edgecolor='Black',palette=['#3455db',"#e4931b"])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))

sns.countplot(
    x='Transaction_ID',
    data=df,
    order=['High', 'Medium', 'Low'],
    palette='coolwarm',
    edgecolor='black',
    legend=False
)

plt.xlabel("Revenue Category", fontsize=10)
plt.ylabel("Number of Transactions", fontsize=10)
plt.title("Transaction Count by Revenue Category", color='pink', fontsize=20)

plt.tight_layout()
plt.show()
```

```python
# countplot use
plt.figure(figsize=(10,5))
sns.barplot(x='Category',y='Revenue',data=df,edgecolor='black',errorbar=None,hue='Category')
plt.xlabel("Category",fontsize=10)
plt.ylabel("Total Revenue ", fontsize=10)
plt.title("Average total category",color='Pink',fontsize=20)
plt.tight_layout()
plt.show()
```

```python
#Scatter plot .
plt.figure(figsize=(8,5))
sns.scatterplot(x='Price',y='Revenue',data=df,s=180,alpha=0.5,edgecolor='black')
plt.xlabel("Price")
plt.ylabel("Revenue")
plt.title("PRICE VS REVENUE ",fontsize=20)
plt.tight_layout()
plt.show()
```

```python
#relation between quantity and revenue
plt.figure(figsize=(8,5))
sns.scatterplot(x='Quantity',y='Revenue',data=df,s=120,alpha=0.5,edgecolor='blue',hue='Product',style='Category')
plt.xlabel("Quantity")
plt.ylabel("Revenue")
plt.title("Revenue vs  Quantity",fontsize=20,color='pink')
plt.tight_layout()
plt.show()
```

```python
#corr method of corelation
numeric_cols=['Price','Quantity','Revenue','Day','Revenue_Per_Unit']
corelationdata=df[numeric_cols].corr()
print("Corelation Matrix:\n",corelationdata.round(2))
```

```python
plt.figure(figsize=(20,8))
sns.heatmap(corelationdata,annot=True,cmap='coolwarm',linewidth=1,linecolor='pink',square=True)
plt.title("CORELATION BETWEEN DATA",fontsize=20,color='pink')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))

sns.countplot(
    x='Is_High_Value',
    data=df,
    palette='Set2',
    edgecolor='black',
    hue='Category'
)

plt.title("High Value vs Not High Value Transactions", fontsize=16)
plt.xlabel("Transaction Type", fontsize=12)
plt.ylabel("Number of Transactions", fontsize=12)

plt.tight_layout()
plt.show()
```
